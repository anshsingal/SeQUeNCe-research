```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gpu_file = open("gpu_data.csv", 'r')
data = []
for i in gpu_file.readlines():
    i = i.split(",")
    if i[0] != "Num bits":
        data.append([int(i[0]), float(i[1][:-2])])
data = sorted(data, key = lambda x:x[0])
gpu_data = list(map(list, zip(*data)))

cpu_file = open("cpu_data.csv", 'r')
data = []
for i in cpu_file.readlines():
    i = i.split(",")
    if i[0] != "Num bits":
        data.append([int(i[0]), float(i[1][:-2])])
data = sorted(data, key = lambda x:x[0])
cpu_data = list(map(list, zip(*data)))

plt.plot(gpu_data[0], gpu_data[1], "-x", label = "GPU")
plt.plot(cpu_data[0], cpu_data[1], "-o", label = "CPU")
plt.yscale('log')
plt.xscale('log')
plt.xlabel("Number of classical symbols")
plt.ylabel("Processing time")
plt.title("CPU vs GPU processing times")
plt.legend()
```
